Function for work done to a particular task :


```python

def findworkdone(proj_name , issue_id):
    if issue_id == "ALL":
        filtered = df[df['Project_Key'] == proj_name]
        if not filtered.empty:
            return filtered['WorkDone'].sum() / len(filtered)
        else:
            return 404 
        
    r = df[(df['Issue_ID']==issue_id) & (df['Project_Key']==proj_name)]
    if not r.empty:
          
          return r['WorkDone'].values[0]
    else:
         return 404
   
```

```python
df1 = df[(df['Status'].isin(['In Progress', 'Blocked', 'To Do'])) & (df['Days_remaining'].dt.days > 0)]
```

```python
df1.sample(5)
```

```python
df1['Workleft']=100-df['WorkDone']
```

```python
df1.sample(5)
```

```python
df1['Days_Spent']=df1['days_provided_to_complete']-df1['Days_remaining'].dt.days
```

```python
df1.sample(2)
```

```python
df1['Risky'] = 0

for i, row in df1.iterrows():
    days_remaining = row['Days_remaining'].days if pd.notnull(row['Days_remaining']) else 0
    if row['Days_Spent'] > 0:
        projected_work = (row['WorkDone'] / row['Days_Spent']) * days_remaining
    else:
        projected_work = 0

    if projected_work < row['Workleft']:
        df1.at[i, 'Risky'] = 1
  
```
